Combining Data in Pandas (Left,Right,Inner Joins and Union)

```python
import pandas as pd
```

```python
#Import & process csv files and display 1st dataset

read_table1=pd.read_csv(r"D:\GE Data Analytics Data Set\ge task 1 for tableau\av_engine_data_aic.csv")
read_table1.head(3)
```

```python
#Read and display the 2nd dataset 

read_table2=pd.read_csv(r"D:\GE Data Analytics Data Set\ge task 1 for tableau\av_engine_data_axm.csv")
read_table2.head(3)
```

```python
#Read and display the 3rd dataset 

read_table3=pd.read_csv(r"D:\GE Data Analytics Data Set\ge task 1 for tableau\av_engine_data_fron.csv")
read_table3.head(3)
```

```python
#Merge datasets (1 & 2) by left join

lft_jn=read_table1.merge(read_table2,on='destination_icao',how='left',suffixes=('_AIC', '_AXM'))
lft_jn.head()
```

```python
# Merge datasets(1 & 2) by right join

lft_jn2=read_table1.merge(read_table2,on='destination_icao',how='right', suffixes=('_AIC', '_AXM'))
lft_jn2.head(9)
```

```python
#Display columns for the left join table

lft_jn2.columns
```

```python
#Drop first thirty-one of the columns of the left join table

lft_jn2.drop(lft_jn2.iloc[:,0:32], inplace=True, axis=1)
print(lft_jn2.head())
```

```python
# Merge tables(1 & 2) by inner join

lft_jn3=read_table1.merge(read_table2,on='destination_icao',how='inner',suffixes=('_AIC', '_AXM'))
lft_jn3.head()
```

```python
#Generating union for the three structured av_engine_data sets
# Append the three datasets into one table

data_sets=[read_table1,read_table2,read_table3]
disply_reslt=pd.concat(data_sets)
disply_reslt
```
